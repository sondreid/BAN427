```python
sns.set_theme(palette='pastel')
def bar_plot(df, x_var, hue_var, y_var, label_title, x_label, y_label, x_axis_label, y_axis_label):
    """
    Generates a bar plot with hue.
    Parameters:
        df: input dataframeø7609
        x_var : x variable
        hue_var: category variable. Is left out if empty string "". 
        y_var: y variable
        label_title: Title of categories
        x_label:
    """ 
    if hue_var == "":
        ax = sns.barplot(data = df,
                x = x_var, 
                y = y_var)
        ax.set_ylabel(y_axis_label)
        ax.set_xlabel(x_axis_label)
    else: 
        ax = sns.barplot(data = df,
                    x = x_var, 
                    y = y_var, 
                    hue = hue_var)
        ax.set_ylabel(y_axis_label)
        ax.set_xlabel(x_axis_label)
        labels = [x_label, y_label]
        h, l = ax.get_legend_handles_labels()
        ax.legend(h, labels, title = label_title)
    return ax
```

```python

# Creating features and prediction variables
x   = df.loc[:, ~df.columns.isin(['TIME1', 'NUMBER_COVERS_TIME2', 
'PREMIUM_INCREASE', 'TIME2', 'TOTAL_PREM_TIME2', 'AVERAGE_INCOME_COUNTY_TIME1','TENURE_TIME2','FULL_CHURN', 'PARTIAL_CHURN', 'MORE_SALE'])]



## One hot encoding age group feature
one_hot_encoder = OneHotEncoder(handle_unknown= 'ignore')

one_hot_encoder.fit(df[['AGE_GROUP']])

age_group_hot = one_hot_encoder.transform(df[['AGE_GROUP']]).toarray()
age_group_hot

age_group_hot_df = pd.DataFrame(age_group_hot)
age_group_hot_df.columns = one_hot_encoder.get_feature_names()

# Concat into x
x = pd.concat([x.reset_index(drop=True), age_group_hot_df], axis=1)
x = x[x.columns.difference(['AGE', 'AGE_GROUP'])] # Remove string AGE_GROUP


# Check for NaN
x[x.isna().any(axis=1)]


y_full_churn    = df['FULL_CHURN']
y_partial_churn = df['PARTIAL_CHURN']
y_more_sale     = df['MORE_SALE']



def variable_importance(model):
    """"
    Function that takes a fitted model and enumerates its features' importance.
    Prints a barplot
    parameters:
        @model: fitted model
    """
    importance = model.coef_[0]
    for i,v in enumerate(importance):
            print('Feature: %0d, Score: %.5f' % (i,v))
    
    pyplot.bar([x for x in range(len(importance))], importance)
    pyplot.show()
    

def roc(ytrain, x_train, ytest, x_test, model, title):
    """
    Function that draws a ROC curve based o
    Parameters:
        @ytrain: target feature training set
        @x_train: features training set
        @y_test: target feature test set
        @x_test: features test set
        @model: a fitted model
    """
    fit_proba = model.predict_proba(x_train)
    yprob_pred = model.predict_proba(x_test)

    fpr, tpr, tr = metrics.roc_curve(ytest, yprob_pred[:,1])
    auc = metrics.roc_auc_score(ytest, yprob_pred[:, 1])

    fpr1, tpr1, tr = metrics.roc_curve(ytrain, fit_proba[:,1])
    auc1 = metrics.roc_auc_score(ytrain, fit_proba[:,1])

    plt.figure(num = None, figsize = (10,10), dpi = 80)
    plt.plot((1,0), (1,0), ls = "--", c = ".3")
    plt.xlabel('False positive rate')
    plt.plot(fpr, tpr, label = 'Test data')
    plt.plot(fpr1, tpr1, label = 'Train data')
    plt.ylabel('True positive rate')
    plt.title(title)
    plt.legend()
    plt.show()
    return plt




def clear_conf_matrix(test_set, prediction_set, title):
  """ 
  Creates and renders a confusion matrix 
  Parameters:
    @test_set: dataset containing test data
    @prediction_set: dataset containing predictions produced by model
    @title: String, title in rendered plot
  """
  matrix = confusion_matrix(test_set, prediction_set)
  matrix_display = ConfusionMatrixDisplay(matrix, display_labels=['0','1'])
  matrix_display.plot(cmap='BuGn', values_format = '.0f')
  matrix_display.ax_.set(
                title= title)
  plt.show()
```

# Comparing models

### Determining optimal treshold

```python
### Optimal threshold

def find_optimal_threshold(model, test_set_x, test_set_y):
    """
    Outputs .....
    """
    preds_model = model.predict_proba(test_set_x)[:,1]
    fpr, tpr, thresholds = metrics.roc_curve(test_set_y, preds_model)
    gmeans = sqrt(tpr * (1-fpr))
    ix = argmax(gmeans)
    print('Best Threshold=%f, gmeans =%.3f' % (thresholds[ix], gmeans[ix]))
    return thresholds[ix]
```

```python
df_avg_income = df[df['AVERAGE_INCOME_COUNTY_TIME1'].notna()] 

y_full_churn_avg_income   = df_avg_income['FULL_CHURN']

df_avg_income   = df_avg_income.loc[:, ~df_avg_income.columns.isin(['TIME1', 'NUMBER_COVERS_TIME2', 'PREMIUM_INCREASE', 'TIME2', 'TOTAL_PREM_TIME2','TENURE_TIME2','FULL_CHURN', 'PARTIAL_CHURN', 'MORE_SALE'])]


#One-hot encoding
one_hot_encoder_avg_income = OneHotEncoder(handle_unknown= 'ignore')

one_hot_encoder_avg_income.fit(df_avg_income[['AGE_GROUP']])
df_avg_income_one_hot = one_hot_encoder_avg_income.transform(df_avg_income[['AGE_GROUP']]).toarray()
df_avg_income_one_hot = pd.DataFrame(df_avg_income_one_hot)
df_avg_income_one_hot.columns = one_hot_encoder_avg_income.get_feature_names()

# Concat into x_avg_income
x_avg_income = pd.concat([df_avg_income.reset_index(drop=True), df_avg_income_one_hot], axis=1)
x_avg_income = x_avg_income[x_avg_income.columns.difference(['AGE', 'AGE_GROUP'])] # Remove string AGE_GROUP

# Remove NA's from income variable
x_avg_income = x_avg_income[x_avg_income['AVERAGE_INCOME_COUNTY_TIME1'].notna()]


# Split into training and test set
xtrain_avg_income_fc, xtest_avg_income_fc, ytrain_avg_income_fc, ytest_avg_income_fc = train_test_split(x_avg_income, y_full_churn_avg_income, test_size = 0.2, random_state = 0)




```

### Training new models: Average county income

```python
# Randomforest
rfc_avg_income_fc = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
rfc_avg_income_fc.fit(xtrain_avg_income_fc, ytrain_avg_income_fc)

# Linear Discriminant Analysis
lda_avg_income_fc = LinearDiscriminantAnalysis()
lda_avg_income_fc.fit(xtrain_avg_income_fc, ytrain_avg_income_fc)


# Gradient boosting machine
gbm_avg_income_fc = GradientBoostingClassifier()
gbm_avg_income_fc.fit(xtrain_avg_income_fc, ytrain_avg_income_fc)

# Naive bayes
nb_avg_income_fc = GaussianNB()
nb_avg_income_fc.fit(xtrain_avg_income_fc, ytrain_avg_income_fc)


# Logistic regression model
logreg_avg_income_fc = LogisticRegression(random_state = 0)
logreg_avg_income_fc.fit(xtrain_avg_income_fc, ytrain_avg_income_fc)
```

### Evaluate models 

```python
## Naive Bayes 
ypred_nb_avg_income = (nb_avg_income_fc.predict_proba(xtest_avg_income_fc)[:,1]  >= find_optimal_threshold(nb_avg_income_fc, xtest_avg_income_fc, ytest_avg_income_fc)).astype(bool)
print('RECALL NB: %.3f' % recall_score(ytest_avg_income_fc, ypred_nb_avg_income))
print('Accuracy NB: %.3f' % accuracy_score(ytest_avg_income_fc, ypred_nb_avg_income))
clear_conf_matrix(ytest_avg_income_fc, ypred_nb_avg_income, 'Confusion matrix Naive Bayes prediction model')
roc(ytrain_avg_income_fc, xtrain_avg_income_fc, ytest_avg_income_fc, xtest_avg_income_fc, nb_avg_income_fc, 'ROC Naive Bayes')



## Logistic regression
ypred_logreg_avg_income = (logreg_avg_income_fc.predict_proba(xtest_avg_income_fc)[:,1]  >= find_optimal_threshold(logreg_avg_income_fc, xtest_avg_income_fc, ytest_avg_income_fc)).astype(bool)
print('RECALL logreg: %.3f' % recall_score(ytest_avg_income_fc, ypred_logreg_avg_income))
print('Accuracy logreg: %.3f' % accuracy_score(ytest_avg_income_fc, ypred_logreg_avg_income))
clear_conf_matrix(ytest_avg_income_fc, ypred_logreg_avg_income, 'Confusion matrix logreg prediction model')
roc(ytrain_avg_income_fc, xtrain_avg_income_fc, ytest_avg_income_fc, xtest_avg_income_fc, logreg_avg_income_fc, 'ROC logreg')




### Random forest

ypred_rfc_avg_income = (rfc_avg_income_fc.predict_proba(xtest_avg_income_fc)[:,1] > find_optimal_threshold(rfc_avg_income_fc, xtest_avg_income_fc, ytest_avg_income_fc)).astype(bool)
print('RECALL RF: %.3f' % recall_score(ytest_avg_income_fc, ypred_rfc_avg_income))
print('Accuracy RF: %.3f' % accuracy_score(ytest_avg_income_fc, ypred_rfc_avg_income))
clear_conf_matrix(ytest_avg_income_fc, ypred_rfc_avg_income, 'Confusion matrix Random Forest prediction model')
roc(ytrain_avg_income_fc, xtrain_avg_income_fc, ytest_avg_income_fc, xtest_avg_income_fc, rfc_avg_income_fc, 'ROC Random Forest ')


### LDA

ypred_lda_avg_income = (lda_avg_income_fc.predict_proba(xtest_avg_income_fc)[:,1]  >= find_optimal_threshold(lda_avg_income_fc, xtest_avg_income_fc, ytest_avg_income_fc)).astype(bool)
print('RECALL LDA: %.3f' % recall_score(ytest_avg_income_fc, ypred_lda_avg_income))
print('Accuracy LDA: %.3f' % accuracy_score(ytest_avg_income_fc, ypred_lda_avg_income))
clear_conf_matrix(ytest_avg_income_fc, ypred_lda_avg_income, 'Confusion matrix Linear Discriminat Analysis prediction model')
roc(ytrain_avg_income_fc, xtrain_avg_income_fc, ytest_avg_income_fc, xtest_avg_income_fc, lda_avg_income_fc, 'LDA GBM ')



#GBM
ypred_gbm_avg_income = (gbm_avg_income_fc.predict_proba(xtest_avg_income_fc)[:,1]  >= find_optimal_threshold(gbm_avg_income_fc, xtest_avg_income_fc, ytest_avg_income_fc)).astype(bool)
print('RECALL gbm: %.3f' % recall_score(ytest_avg_income_fc, ypred_gbm_avg_income))
print('Accuracy gbm: %.3f' % accuracy_score(ytest_avg_income_fc, ypred_gbm_avg_income))
clear_conf_matrix(ytest_avg_income_fc, ypred_gbm_avg_income, 'Confusion matrix Naive Bayes prediction model')
roc(ytrain_avg_income_fc, xtrain_avg_income_fc, ytest_avg_income_fc, xtest_avg_income_fc, gbm_avg_income_fc, 'ROC GBM ')

```
